```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('heart.csv')
data
```

```python
data.isnull().sum()
```

since no values are left blank || no missing data we can proceed to segregation

['age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','oldpeak', 'slope','ca','thal'] are the cols which are the inputs form the user <br />
And ['target'] is the output needed to be processed

```python
x = data[['age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','oldpeak', 'slope','ca','thal']] 
y = data[['target']]
```

We can see that the input and output datas are segregated <br /> But we acn observe that the values in the age col is 2 digit and other vals in other cols are single digit hence we need to consider the feature scaling

```python
from sklearn.preprocessing import StandardScaler
Scaler = StandardScaler()
# Scaler.fit(x)
# #x = Scaler.transform(x)
```

Spliting the data into test and train data in the ratio 1:4

```python
from sklearn.model_selection import train_test_split
import numpy as np
X_train, X_test, y_train, y_test = train_test_split(x, y, train_size=0.8, random_state=42)
```

```python
X_train
```

```python
X_test
```

Clearly 205/820 == 1:4

```python
Scaler.fit(X_train)
X_train = Scaler.transform(X_train)
Scaler.fit(X_test)
X_test = Scaler.transform(X_test)
```

```python
X_test
```

```python
X_train
```

Clearly, we can observe that the data is scaled and the data in age is not being prioritiesed <br /> Now we can load and train the model

```python
from sklearn.svm import SVC
```

```python
# Create an SVM model
svm_model = SVC(kernel='linear', C=1)

# Train the model on your data
svm_model.fit(X_train, y_train)
```

```python
# Make predictions on your test data
predictions = svm_model.predict(X_test)
```

```python
# Calculate accuracy
accuracy = svm_model.score(X_test, y_test)
print(f"Model Accuracy: {accuracy}")
```
